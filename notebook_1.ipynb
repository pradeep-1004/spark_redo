```python
df = spark.read.format('csv')\
        .option('inferSchema','true')\
        .option('header','true')\
        .load('/Volumes/spark_redo/spark/bronze_data/BigMart Sales.csv')
```

```python
display(df.limit(5))
```

# **transformation**

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
df.select(col('Item_Identifier'),col('Item_Fat_Content'),col('Item_Visibility')).display()
```

```python
df.select(col('Item_Identifier').alias('Item_ID')).display()
```
