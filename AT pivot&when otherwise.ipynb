```python
df = spark.read.format('csv')\
    .option('header', 'true')\
    .option('inferSchema', 'true')\
    .load('/Volumes/spark_redo/spark/bronze_data/BigMart Sales.csv')


display(df.limit(10))
```

# **pivot**

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
df.groupBy('Item_Type').pivot('Outlet_Type').agg(sum("Item_MRP")).display()

```

# **when otherwise**

```python
from pyspark.sql.functions import when, col

df = df.withColumn('healthy',when(col('Item_Fat_Content') == 'Regular','healthy').otherwise('unHealthy'))
display(df)
```
