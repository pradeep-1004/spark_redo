```python
df = spark.read.format('csv')\
    .option('header', 'true')\
    .option('inferSchema', 'true')\
    .load('/Volumes/spark_redo/spark/bronze_data/BigMart Sales.csv')
```

# **handling nulls**

```python
df.dropna('any').display()
```

```python
df.dropna(subset=['item_Weight']).display()
```

# **filling nulls**

```python
df.fillna('NA').display()
```

```python
df.fillna({'item_Weight': 0}).display()
```

# **split**

```python
from pyspark.sql.functions import *
df.withColumn('Outlet_Type',split('Outlet_Type',' ')).display()
```

```python
df.withColumn('Outlet_Type',split('Outlet_Type',' ')[1]).display()
```

# **explode**

```python
from pyspark.sql.functions import *
```

```python
df_exp = df.withColumn('Outlet_Type',split('Outlet_Type',' '))

display(df_exp)
```

```python
df_exp = df_exp.withColumn('Outlet_Type',explode('Outlet_Type'))

display(df_exp)
```
