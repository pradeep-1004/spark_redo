```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
df = spark.read.format('csv')\
        .option('header', 'true')\
        .option('inferSchema', 'true')\
        .load('/Volumes/spark_redo/spark/bronze_data/BigMart Sales.csv')
```

```python
display(df.limit(10))
```

```python
df.printSchema()
```

### select

```python
df.select(col('Item_Identifier'),col('Item_Type')).display()
```

### **** alias

```python
df.select(col('Item_Weight').alias('item_in_kg')).display()
```

### **filter**

```python
df.filter(col("Item_Fat_Content")=='Regular').display()
```

```python
df.filter((col('Item_Fat_Content')=='Regular') & (col('Item_Type')=='Dairy')).display()
```

```python
df.filter((col('Outlet_Size').isNull()) & (col('Outlet_Location_Type').isin('Tier 1','Tier 2'))).display()
```

### with column **renamed**

```python
df.withColumnRenamed('Item_MRP','item_inr').display()
```

# **with column**

```python
df.withColumn('total_price',col("Item_MRP")*col("Item_Weight")).display()
```

```python
df = df.withColumn('Item_Fat_Content',regexp_replace(col('Item_Fat_Content'),"Regular","Reg"))\
    .withColumn('Item_Fat_Content',regexp_replace(col('Item_Fat_Content'),"Low Fat","Lf"))

df.display()
```

# **type casting**

```python
df_type = df.withColumn('item_Weight',col("Item_Weight").cast(StringType()))
```

```python
df_type.printSchema()
```

# **sort**

```python
df.sort('Outlet_Establishment_Year').display()
```

```python
df.sort(col('Outlet_Establishment_Year').desc()).display()
```

```python
df.sort(["Item_Identifier",'Item_Weight'],ascending = [0,0]).display()
```

```python
df.sort(["Item_Identifier",'Item_Weight'],ascending = [0,1]).display()
```

# **limit**

```python
df.limit(5).display()
```

# **drop**

```python
df.drop("Item_Fat_Content").display()
```

```python
df.drop("Item_Fat_Content",'Item_Identifier').display()
```
