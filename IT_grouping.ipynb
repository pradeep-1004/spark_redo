```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
df = spark.read.format('csv')\
        .option('header', 'true')\
        .option('inferSchema', 'true')\
        .load('/Volumes/spark_redo/spark/bronze_data/BigMart Sales.csv')
```

```python
display(df)
```

```python
df.groupBy('Item_Type').agg(sum("Item_MRP")).display()
```

```python
df.groupBy('Item_Type').agg(avg("Item_MRP")).display()
```

```python
df.groupBy('Item_Type','Outlet_Size').agg(sum("Item_MRP").alias('in_inr')).display()
```

```python
df.groupBy('Item_Type','Outlet_Size').agg(sum("Item_MRP").alias('avg_inr'),avg("Item_MRP").alias('avg_inr')).display()
```

# **collect list**

```python
data = [('user1','book1'),
        ('user1','book2'),
        ('user2','book2'),
        ('user2','book4'),
        ('user3','book1')]

schema = 'user string, book string'

df_book = spark.createDataFrame(data,schema)

df_book.display()
```

```python
df_book.groupBy('user').agg(collect_list("book")).display()
```
